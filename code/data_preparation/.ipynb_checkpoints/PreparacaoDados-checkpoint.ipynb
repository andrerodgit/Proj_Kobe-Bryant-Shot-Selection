```python
# Importar Bibliotecas Necessárias
import os
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
import mlflow
import seaborn as sns
import matplotlib.pyplot as plt

# Definir a função para pré-processamento de dados
def preprocess_data(raw_data_file):
    """
    Função para pré-processamento e filtragem de dados.
    
    Args:
    - raw_data_file (str): Caminho do arquivo CSV contendo os dados brutos.
    
    Returns:
    - processed_data (DataFrame): DataFrame pré-processado e filtrado.
    """
    # Carregar os dados brutos do arquivo CSV
    df_raw = pd.read_csv(raw_data_file, sep=',')
    
    # Dimensão do DataFrame filtrado original
    print("Dimensão do DataFrame original:", df_raw.shape)
    print("Head do DataFrame original:")
    print(df_raw.head())

    print("##############################################################")

    # Verificar dados faltantes no dataset original
    print("Dados faltantes no dataset original:")
    print(df_raw.isnull().sum())

    # Remover linhas com dados faltantes
    df_raw.dropna(inplace=True)

    print("##############################################################")

    # Filtrar os dados onde o valor de shot_type é igual a '2PT Field Goal'
    df_filtered = df_raw[df_raw['shot_type'] == '2PT Field Goal']

    # Filtrar o DataFrame para incluir apenas as colunas especificadas
    columns_to_keep = ['lat', 'lon', 'minutes_remaining', 'period', 'playoffs', 'shot_distance', 'shot_made_flag']
    processed_data = df_filtered[columns_to_keep].copy()

    # Substituir valores faltantes na coluna 'shot_made_flag' por 0 (indicando que o arremesso foi errado)
    processed_data['shot_made_flag'] = processed_data['shot_made_flag'].astype(int)

    # Dimensão do DataFrame filtrado e pré-processado
    print("Dimensão do DataFrame filtrado e pré-processado:", processed_data.shape)
    
    return processed_data


# Definir diretório de execução do MLflow
mlflow.set_tracking_uri("../../data/modeling/tracking/mlruns")

# Verificar se o experimento existe, senão, criar
mlflow.set_experiment("PreparacaoDados")

# Inicie uma rodada do MLflow
with mlflow.start_run(run_name="PreparacaoDados"):
    
    # Caminho do arquivo CSV contendo os dados brutos
    raw_data_file = "../../data/raw/data.csv"
    
    # Executar a função de pré-processamento de dados
    df_kobe = preprocess_data(raw_data_file)

    # Salvar o DataFrame filtrado
    data_file = "../../data/processed/data_filtered.parquet"
    df_kobe.to_parquet(data_file)
   
    # Salvar o DataFrame pré-processado em arquivos Parquet de desenvolvimento e produção
    output_file_dev = "../../data/raw/dataset_kobe_dev.parquet"
    output_file_prod = "../../data/raw/dataset_kobe_prod.parquet"
    df_kobe.to_parquet(output_file_dev)
    df_kobe.to_parquet(output_file_prod)
    
    # Log dos arquivos de dados processados como artefatos no MLflow
    mlflow.log_artifact(output_file_dev)
    mlflow.log_artifact(output_file_prod)

    print("Arquivos Parquet de dados brutos de prod e dev criados com sucesso.")
    
    # Calcular dimensão do dataset resultante
    df_dimension = df_kobe.shape[0]
    
    # Log da dimensão do dataset resultante
    mlflow.log_param("dimensao_dataset", df_dimension)

# Divisão dos Dados em Conjuntos de Treino (80%) e Teste (20 %)
def split_data(data, test_size=0.2):

    X = data.drop('shot_made_flag', axis=1)
    y = data['shot_made_flag']
    
    # Dividir os dados de forma aleatória e estratificada
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, stratify=y, random_state=42)
    
    return X_train, X_test, y_train, y_test

# Armazenar os datasets resultantes em arquivos Parquet
def save_data(X, y, file_path):

    # Criar o diretório se não existir
    os.makedirs(os.path.dirname(file_path), exist_ok=True)
    
    # Concatenar os recursos e o alvo
    data = pd.concat([X, y], axis=1)
    
    # Salvar os dados em formato Parquet
    data.to_parquet(file_path, index=False)

# Dividir os dados em conjuntos de treino e teste
X_train, X_test, y_train, y_test = split_data(df_kobe)

# Exibir informações sobre a divisão
print("Dimensões do conjunto de treino:", X_train.shape, y_train.shape)
print("Dimensões do conjunto de teste:", X_test.shape, y_test.shape)

# Salvar os conjuntos de dados em arquivos Parquet
save_data(X_train, y_train, "../../data/processed/base_train.parquet")
save_data(X_test, y_test, "../../data/processed/base_test.parquet")

# Log dos parâmetros
mlflow.log_param("percentual_teste", 0.2)

# Log das métricas
mlflow.log_metric("tamanho_base_treino", X_train.shape[0])
mlflow.log_metric("tamanho_base_teste", X_test.shape[0])

# Exibir o número absoluto de amostras em cada conjunto
print("Número absoluto de amostras no conjunto de treino:", X_train.shape[0])
print("Número absoluto de amostras no conjunto de teste:", X_test.shape[0])

# Adicionar comentários explicativos da divisão dos dados
comentario = "divisao_dados", "Na divisão de dados, o conjunto de treino contém 80% e o conjunto de teste 20%. Essa divisão foi feita de forma aleatória e estratificada para preservar a distribuição das classes."
mlflow.log_param("divisao_dados", comentario)
print("Divisão de dados: ", comentario)

print("##############################################################")

# Distribuição das Variáveis Numéricas
def plot_numeric_distribution(data):
    fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 10))

    sns.histplot(data['lat'], bins=20, kde=True, ax=axes[0, 0])
    axes[0, 0].set_title('Distribuição de lat')

    sns.histplot(data['lon'], bins=20, kde=True, ax=axes[0, 1])
    axes[0, 1].set_title('Distribuição de lon')

    sns.histplot(data['minutes_remaining'], bins=20, kde=True, ax=axes[0, 2])
    axes[0, 2].set_title('Distribuição de minutes_remaining')

    sns.histplot(data['shot_distance'], bins=20, kde=True, ax=axes[1, 0])
    axes[1, 0].set_title('Distribuição de shot_distance')

    sns.boxplot(x='shot_made_flag', y='lat', data=data, ax=axes[1, 1])
    axes[1, 1].set_title('Distribuição de lat por shot_made_flag')

    sns.boxplot(x='shot_made_flag', y='lon', data=data, ax=axes[1, 2])
    axes[1, 2].set_title('Distribuição de lon por shot_made_flag')

    plt.tight_layout()
    plt.show()

# Chamar a função para plotar as distribuições
plot_numeric_distribution(df_kobe)

print("##############################################################")

# Calcular estatísticas resumidas das variáveis numéricas em relação à variável alvo
summary_stats = df_kobe[['lat', 'lon', 'minutes_remaining', 'shot_distance', 'shot_made_flag']].describe()

# Imprimir estatísticas resumidas
print("Estatísticas Resumidas das Variáveis Numéricas:")
print(summary_stats)

print("##############################################################")

# Calcular média, mediana e desvio padrão das variáveis numéricas para cada valor da variável alvo
numeric_features = ['lat', 'lon', 'minutes_remaining', 'shot_distance']
summary_stats = df_kobe.groupby('shot_made_flag')[numeric_features].agg(['mean', 'median', 'std'])

# Exibir os resultados
print("Relação entre Variáveis Numéricas e Variável Alvo:")
print(summary_stats)

print("##############################################################")

# Calcular a frequência das variáveis categóricas codificadas em relação à variável alvo 'shot_made_flag'
encoded_categorical_variables = [col for col in df_kobe.columns if 'period' in col or 'playoffs' in col]

for var in encoded_categorical_variables:
    # Calcular a frequência de cada categoria
    freq = df_kobe[var].value_counts()
    
    # Calcular a proporção de cada categoria em relação ao total
    proportion = freq / len(df_kobe)
    
    # Exibir os resultados
    print(f"Distribuição da variável '{var}':")
    for category, count in freq.items():
        print(f"- Categoria '{category}': {count} amostras ({proportion[category]:.2%})")
    print()

print("##############################################################")

# Calcular a matriz de correlação entre as variáveis numéricas
correlation_matrix = df_kobe[['lat', 'lon', 'minutes_remaining', 'shot_distance', 'shot_made_flag']].corr()

# Imprimir as informações da matriz de correlação
print("Matriz de Correlação:\n", correlation_matrix)

# Imprimir as correlações com a variável alvo ('shot_made_flag')
print("\nCorrelação com 'shot_made_flag':\n", correlation_matrix['shot_made_flag'])

print("##############################################################")

# Caminho do arquivo CSV contendo os dados brutos
raw_data_file = "../../data/raw/data.csv"
    
# Executar a função de pré-processamento de dados
df_raw = preprocess_data(raw_data_file)

# Calcular estatísticas temporais dos arremessos ao longo do jogo
temporal_stats = df_raw.groupby('minutes_remaining')['shot_made_flag'].mean()

# Imprimir as estatísticas temporais
print("Estatísticas Temporais dos Arremessos ao Longo do Jogo:")
print(temporal_stats)

print("##############################################################")

# Calcular estatísticas descritivas da variável mais relevante (shot_made_flag) para treino e teste
train_target_stats = y_train.describe()
test_target_stats = y_test.describe()

# Imprimir as estatísticas descritivas
print("Estatísticas Descritivas da Variável Mais Relevante (shot_made_flag) para o Conjunto de Treino:")
print(train_target_stats)

print("\nEstatísticas Descritivas da Variável Mais Relevante (shot_made_flag) para o Conjunto de Teste:")
print(test_target_stats)

print("##############################################################")

# Calcular a média da variável 'shot_made_flag' para cada minuto restante no jogo
minutes_remaining_shot_mean = df_kobe.groupby('minutes_remaining')['shot_made_flag'].mean()

print(minutes_remaining_shot_mean)

# Plotar o gráfico de linha
plt.figure(figsize=(10, 6))
plt.plot(minutes_remaining_shot_mean.index, minutes_remaining_shot_mean.values, marker='o', linestyle='-')
plt.title("Taxa de Acerto dos Arremessos em Relação aos Minutos Restantes no Jogo")
plt.xlabel("Minutos Restantes no Jogo")
plt.ylabel("Taxa de Acerto")
plt.grid(True)
plt.show()

print("##############################################################")

# Plotar histograma da distância do arremesso para cada classe (acertou ou não)
plt.figure(figsize=(10, 6))
sns.histplot(df_kobe[df_kobe['shot_made_flag'] == 0]['shot_distance'], color='red', label='Erro')
sns.histplot(df_kobe[df_kobe['shot_made_flag'] == 1]['shot_distance'], color='blue', label='Acerto')
plt.title('Distribuição da Distância do Arremesso por Classe')
plt.xlabel('Distância do Arremesso')
plt.ylabel('Contagem')
plt.legend()
plt.show()

print("##############################################################")

# Plotar densidade da distância do arremesso para cada classe (acertou ou não)
plt.figure(figsize=(10, 6))
sns.kdeplot(df_kobe[df_kobe['shot_made_flag'] == 0]['shot_distance'], color='red', label='Erro', fill=True)
sns.kdeplot(df_kobe[df_kobe['shot_made_flag'] == 1]['shot_distance'], color='blue', label='Acerto', fill=True)
plt.title('Distribuição da Distância do Arremesso por Classe')
plt.xlabel('Distância do Arremesso')
plt.ylabel('Densidade')
plt.legend()
plt.show()

print("##############################################################")

# Plotar histograma com densidade da variável mais relevante para cada conjunto (treino/teste)
plt.figure(figsize=(10, 6))
sns.histplot(y_train, color='green', label='Conjunto de Treino', kde=True, stat='density')
sns.histplot(y_test, color='orange', label='Conjunto de Teste', kde=True, stat='density')
plt.title('Distribuição da Variável Mais Relevante (shot_made_flag) por Conjunto')
plt.xlabel('Valor da Variável Mais Relevante (shot_made_flag)')
plt.ylabel('Densidade')
plt.legend()
plt.show()
```
